```python
import pandas as pds
import numpy as npy
```

```python
name = pds.Series(["Bob", "Amy", "Vicky", "Tim"])
print(name)
print(name.index)
print(name.values) #name的value
```

```python
name = pds.Series(["Bob", "Amy", "Vicky", "Tim"], index=["A", "B", "C", "D"])
print(name)
print(name.index)
print(name[1])
print(name['D'])
```

```python
ind = {"A":"Ant","B":"Bird","C":"Cat"}
animal = pds.Series(ind)
print(animal)
```

```python
name = pds.Series(["Bob", "Amy", "Vicky", "Tim"])
del_name = name.drop(2)
print(name)
print(del_name)
name = pds.Series(["Bob", "Amy", "Vicky", "Tim"], index=["A", "B", "C", "D"])
del_name2 = name.drop("A")
print(del_name2)
```

```python
name = pds.Series(["Bob", "Amy", "Vicky", "Tim"])
name[1] = "Sean"
print(name)
name = pds.Series(["Bob", "Amy", "Vicky", "Tim"], index=["A", "B", "C", "D"])
name["A"] = "Sean"
print(name)
```

```python
name = pds.Series(["Bob", "Amy", "Vicky", "Tim", "Tommy"])
print(name.size)
```

```python
name = pds.Series(["Bob", "Amy", "Vicky", "Tim", "Tommy"])
key = name.str.contains("i")
print(key)
```

```python
name = pds.Series(["Bob", "Amy", "Vicky", "Tim", "Tommy"])
replace_name = name.str.replace("Bob", "Sean")
print(replace_name)
```

```python
numbers = npy.random.randint(0,200,10) #0到199取10個整數值
print(numbers)
N = pds.Series(numbers)
#print(N)
print(N.max())
print(N.min())
print(N.mean())
print(N.sum())
print(N.nlargest(3))
print(N.nsmallest(3))
```

```python
numbers = npy.random.randint(0,200,12)
N = pds.Series(numbers)
print(N)
print(N < 100)
print(N[N < 100])
print(N[N <= 100 ])
print(N[N == 100 ])
print(N[N != 100 ])
print(N[(N >= 30) & (N <=90)])
print(N[N.between(60, 80)])
print(N[(N >= 30) | (N <=90)])
```

```python
numbers = npy.random.randint(0,200,12)
N = pds.Series(numbers)
print(N)
print(N[N.isin(i for i in range(100, 150))])
N[8] = None
print(N)
print(N[N.isnull()])
print(N[N.notnull()])
```
