# 作業: 資料讀寫與繪製圖表

依照以下步驟完成作業:
1. 請讀取 data.csv 檔案中的 x, y 資料並用 matplotlib 畫出資料散佈圖 (使用 scatter 函數)，資料點請用紅色叉叉
2. 將 data.csv 中的 x資料帶入 0.6x+2，得到另一組 y 的資料，利用這組新的 xy 資料在同一張圖上畫出線 (使用 plot 函數)，線條請用藍色
3. 將新產生的xy資料儲存到新的檔案，檔名叫做 "data_new.csv"

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('data.csv')
x = data['x']
y = data['y']

plt.scatter(x, y, color='red', marker='x', label='Original Data')
new_y = 0.6 * x + 2
plt.plot(x, new_y, color='blue', label='New Line')


new_data = pd.DataFrame({'x': x, 'y': new_y})
new_data.to_csv('data_new.csv', index=False)

plt.xlabel('X')
plt.ylabel('Y')
plt.legend()

plt.show()
```
