### 2. 定義一個函數，輸入兩個1~9的正數a與b，輸出axb到bxa之間的99乘法字串 (20分)
舉例: 輸入 5, 6 ，輸出 5x6=30, 5x7=35 ... 6x4=24, 6x5=30

```python
def list_99(a, b):
    str_99 = ''
    for i in range(a, b+1):
        if i==a:
            for j in range(b, 10):
                str_99 += str(i)+'x'+str(j)+'='+str(i*j)+'\n'
        elif i==b:
            for j in range(1, a+1):
                str_99 += str(i)+'x'+str(j)+'='+str(i*j)+'\n'
        else:
            for j in range(1, 10):
                str_99 += str(i)+'x'+str(j)+'='_+str(i*j)+'\n'    
    return str_99
```

```python
print(list_99(3,6))
```

### 3. 定義一個函數，輸入一個正整數，回傳一個list，內為小於等於該正整數的所有質數 (20分)
舉例: 輸入11，輸出[1, 2, 3, 5, 7, 11]

```python
def get_prime_numbers(x):
    pnums = [i for i in range(1, x+1)]
    return pnums
```

```python
print(get_prime_numbers(13))
```
